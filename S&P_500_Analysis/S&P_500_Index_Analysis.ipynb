# S&P 500 Index Investment Analysis

<h3>Description</h3>

The below analysis will explore the S&P 500 Index's closing prices within a four-decade time frame. The objective is to analyze the index and determine if the index is a valuable long-term investment.

Four decades:

Decade 1: 1979 - 1989<br>
Decade 2: 1989 - 1999<br>
Decade 3: 1999 - 2009<br>
Decade 4: 2009 - 2019<br>

The analysis will seek to understand the compounded returns (not including dividends) the index has produced within the above time frames. 

A/B tests will be used to infer if the difference in means of the S&P 500 Index closing prices between decades is statistically significant. The goal is to observe if the closing prices are higher or lower between the time frames and determine the investment risk. 

If the prices are higher in each new decade and statistically significant, intuitively, this will be determined to be a positive sign and that the index is producing profits; the opposite is true for the opposite results.

In addition to the above, the analysis will observe other macro and micro economic data trends regarding real U.S GDP, S&P 500 Index earnings per share (EPS), average U.S 10-year bond yields, and average annual U.S inflation rates to aid in the analysis.

The analysis can be taken advantage of by any individual or entity looking for a valuable investment opportunity.

<h3>Trend Insights</h3>

<h4>S&P 500 Index Daily Closing Prices vs Years</h4><br>
The S&P 500 Index daily closing prices have increased within the four decades measured (1979-2019) and have grown annually at a compound rate of 8% (dividends not reinvested). However, the slope is not a consistent rise due to volatile periods in the mid-1990s through 2010 and can be explained by the dot-com bubble in 2001 and the financial crisis in 2007-2009.

Earlier recessions took place in 1980 - 1982 due to the energy crisis (high oil prices, high inflation, high interest rates) and a recession in 1990 - 1991 (Federal Reserve Bank increasing interest rates to battle high inflation).

The S&P 500 Index has grown and provided solid returns via price increases, even having gone through significant adverse economic events. Therefore, being invested in the S&P 500 Index over the long term appears to be a profitable investment through the volatile time period.  

<h4>Real U.S GDP vs Years</h4><br>
Real U.S GDP has grown steadily over the four decades measured and has grown at a compound rate of 3% annually. However, there are flat lines and dips through the recession timelines mentioned above. 

Additionally, we can observe the pricing of the S&P 500 Index as adverse economic events took place. For example, while the effects of the early 1980s may seem small in the line plots above, this period was known to be the worst recession since the great depression (1929-1939). During this period, however, prices seemed to remain steady.

The recession during 1990 - 1991 is not very apparent in the S&P 500 Index's daily closing prices.

Looking forward to the dot-com bubble and the financial crisis, stock prices became very volatile.

If the investor invested in the S&P 500 index and remained unphased by the volatility or adverse economic events (invested capital was not needed to finance everyday life and little to no debt), the investor would have capitalized on the resiliency and strength of the American economy. The investor would have profited from the investment with a high margin of safety.
<br>

<h4>S&P 500 Index EPS vs Years</h4><br>
The S&P 500 Index EPS has to grow within the four decades measured. There is more volatility in comparison to annual real U.S GDP and the S&P 500 Index's daily closing prices. 

The earnings on a per-share basis have been quite volatile; however, the annual compounded rate of growth for the EPS is 3% annually (in line with Real U.S GDP). This continues to build the case that the longer time horizon increases profits through volatile events and that the S&P 500 Index is a profitable investment over time.
<br>

<h4>Average U.S 10-Year Bond Yields vs Years</h4><br>
Average U.S 10-Year Bond Yields have shown to decline steadily over the four decades. The annual compound rate of bond yields is -4% annually. 

Due to the known inverse relationship asset prices have with interest rates, the steady drop in rates is a catalyst for increased S&P 500 Index daily closing prices; however, this cannot be the only reason due to the S&P 500 index has shown increasing earnings (EPS).
<br>

<h4>Average Annual U.S Inflation vs Years</h4><br>
The Average Annual U.S Inflation has declined steadily over the four decades measured and has declined annually at a compound rate of -4%. This appears to be the reason that interest rates have not increased, in addition to the recessions that had them lowered.
<br>

<h3>S&P 500 Index Daily Closing Prices Insights for Decade 1 and Decade 2</h3>

```python
# T-tests between Decade 1 and Decade 2
series_stats_gen(
    string = 'Decade 1', 
    column = da_1['s&p_closing_prices'], 
    
    string2 = 'Decade 2',
    column2 = da_2['s&p_closing_prices'], 
    
    title = 'Decade 1 Prices vs Decade 2 Prices', 
    y_axis = 'S&P 500 Index Daily Closing Prices', 
    
    compound_years = 20, 
    length = 10, 
    width = 5,
    font = 1.5
)
```

The null hypothesis stating that there is no difference between the means of the S&P 500 Index daily closing prices for Decade 1 and Decade 2 is rejected. 

The prices between Decade 1 and Decade 2 are statistically significant at p < 0.05. With 95% confidence, the difference can be found between -386.29 & -366.81 (distribution: 166.9, 176.64, 186.38). This shows that prices were lower in Decade 1 in comparison to Decade 2 and therefore, on average, the investor would have turned a profit if the investor was invested in the S&P 500 Index during Decade 1 through Decade 2.

Additionally, if The investor bought the S&P 500 Index at the beginning of Decade 1 and sold at the end of Decade 2 (20 years), the investor would have earned an actual compounded return of 14%.

If the investor invested in Decade 1 through Decade 2, the investor would have earned satisfactory returns with a high margin of safety.

<h3>Conclusion</h3>

In conclusion, the S&P 500 Index has proven to be a profitable investment with a high margin of safety for the long-term investor. 

The investor should consistently deploy capital in the investment and, as a result, will underpay during adverse economic periods and overpay during good economic periods. The advantage of buying cheap and the disadvantage of buying expensive will ultimately cancel each other out. 

Incidentally, if the investor tries to time the market, the investor may risk continuously overpaying for the investment and miss good buying opportunities at cheaper prices. On the contrary, the investor may get lucky and be able to capitalize; however, this may not be reproducible, and the investor may face opportunity costs by not being invested.
